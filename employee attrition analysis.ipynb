```python
#getting object column names to convert it into numerical column
def catcolumns(df):
    return [cname for cname in df.columns if df[cname].dtypes=='object' ]

catcolumns(tdf4)
```

```python
def toperformencoding(df,columns_to_encode):
    return pd.get_dummies(df,columns=columns_to_encode)
cat_cols=catcolumns(tdf4)
tdf6=toperformencoding(tdf4,cat_cols)
```

```python
tdf6
```

```python

hr=tdf6.columns.values.tolist()
y ='status'
x=[cname for cname in hr if cname not in y]
```

```python
x
```

```python
tdf6.drop('Emp ID',axis=1,inplace=True)
```

### Feature Selection
The Recursive Feature Elimination (RFE)

```python
from sklearn.feature_selection import RFE
from sklearn.linear_model import LogisticRegression

#Recursive Feature Elimination (RFE)
model = LogisticRegression()

rfe = RFE(model, 10)
rfe = rfe.fit(tdf6[x],tdf6[y])
print(rfe.support_)
print(rfe.ranking_)
```

```python
cols=['Work_accident',
 'time_spend',
 'last_evaluation',
 'number_project',
 'satisfaction_level',
 'dept_sales',
 'salary_high',
 'salary_low',
 'salary_medium']
x=tdf6[cols]
y=tdf6['status']
```

```python
from sklearn.model_selection import train_test_split
xtrain,xtest,ytrain,ytest=train_test_split(x,y,test_size= 0.3,random_state=0)
```

```python
from sklearn.metrics import accuracy_score
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import confusion_matrix
from sklearn.metrics import classification_report
rf=RandomForestClassifier()
rf.fit(xtrain,ytrain)
pred=rf.predict(xtest)
print(confusion_matrix(ytest,pred))
print(classification_report(ytest,pred))
#print(pred)
```

```python
from sklearn import model_selection
from sklearn.model_selection import cross_val_score
kfold= model_selection.KFold(n_splits=10,random_state=7)
modelCV=RandomForestClassifier()
scoring='accuracy'
results = model_selection.cross_val_score(modelCV, xtrain, ytrain, cv=kfold, scoring=scoring)
```

```python
print(results.mean())
```

```python
from sklearn.svm import SVC
from sklearn.metrics import confusion_matrix
from sklearn.metrics import classification_report
svc = SVC()
svc.fit(xtrain, ytrain)
pred=svc.predict(xtest)
print(confusion_matrix(ytest,pred))
print(classification_report(ytest,pred))
#For SVM
from sklearn import model_selection
from sklearn.model_selection import cross_val_score
kfold = model_selection.KFold(n_splits=10, random_state=7)
modelCV = SVC()
scoring = 'accuracy'
results = model_selection.cross_val_score(modelCV, xtrain, ytrain, cv=kfold, scoring=scoring)
print("10-fold cross validation average accuracy for SVM Classifier: %.3f" % (results.mean()))
```
